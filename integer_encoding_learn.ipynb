## Intger encoding simplelearn 

```python
import numpy as np

docs = [
    'go india',
    'india india',
    'hip hip hurray',
    'jeetega bhai jeetega india jeetega',
    'bharat mata ki jai',
    'kohli kohli',
    'sachin sachin',
    'dhoni dhoni',
    'modi ji ki jai',
    'inquilab zindabad'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token='Devil')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index  # word count dictionary
```

```python
tokenizer.word_counts  # how many time a word came in the docs
```

```python
tokenizer.document_count  # total number of sentences in the docx
```

```python
Sequences = tokenizer.texts_to_sequences(docs)  # to get the sequence of the word 
Sequences
```

```python
from keras.utils import pad_sequences
```

```python
sequences  = pad_sequences(Sequences,padding='post')
sequences
```
