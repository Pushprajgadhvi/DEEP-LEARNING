## DEEP RNN 

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence  import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN,  Dense
```

```python
# load the dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# pad sequences to ensure uniform input size and have the same length 

x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100) 
```

```python
# DEEP RNN model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)), # embeding layer to convert word to vectors 
    SimpleRNN(32, return_sequences=True),                          # RNN layers with 5 units
    SimpleRNN(16),                                                 # another RNN layer with 5 units
    Dense(1, activation='sigmoid')                                 # output layer for binary classification
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
# compile and train the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

```python
# DEEP GRU model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)), # embeding layer to convert word to vectors 
    GRU(32, return_sequences=True),                          # RNN layers with 5 units
    GRU(16),                                                 # another RNN layer with 5 units
    Dense(1, activation='sigmoid')                                 # output layer for binary classification
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
# DEEP LSTM model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM , Dense

model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)), # embeding layer to convert word to vectors 
    LSTM(32, return_sequences=True),                          # RNN layers with 5 units
    LSTM(16),                                                 # another RNN layer with 5 units
    Dense(1, activation='sigmoid')                                 # output layer for binary classification
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```
