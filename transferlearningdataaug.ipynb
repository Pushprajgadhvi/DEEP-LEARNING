## Transfer learning (with the data augmentation)

```python
import tensorflow 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Flatten
from keras.applications.vgg16 import VGG16
```

```python
conv_base = VGG16(
    weights='imagenet',
    include_top = False,
    input_shape=(150,150,3)
)
```

```python
conv_base.summary()
```

```python
model = Sequential()

model.add(conv_base)
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
conv_base.trainable = False
```

```python
batch_size = 32

train_ds = keras.utils.image_dataset_from_directory(
    directory = 'D:/Data Science/Machine Language/deep learning/CNN/dataset/animals/Test',
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 32,
    image_size = (150,150)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory = 'D:/Data Science/Machine Language/deep learning/CNN/dataset/animals/Train',
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 32,
    image_size = (150,150)
)
```

```python
#normalization
def process(image,label):
    image = tensorflow.cast(image/255. ,tensorflow.float32)
    return image,label
train_ds = train_ds.map(process)
validation_ds = validation_ds.map(process)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_ds, epochs = 5 , validation_data=validation_ds)
```

```python
import matplotlib.pyplot as plt

# Create a figure with 2 subplots (rows=1, cols=2)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))


# Plot Accuracy
ax[0].plot(history.history['accuracy'], color='red', label='train')
ax[0].plot(history.history['val_accuracy'], color='blue', label='validation')
ax[0].set_title('Model Accuracy')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Accuracy')
ax[0].legend()

# Plot Loss
ax[1].plot(history.history['loss'], color='red', label='train')
ax[1].plot(history.history['val_loss'], color='blue', label='validation')
ax[1].set_title('Model Loss')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Loss')
ax[1].legend()

plt.tight_layout()
plt.show()
```
