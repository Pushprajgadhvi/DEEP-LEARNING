## Words Vectors in spacy

```python
import spacy

nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp("I am learning Natural Language Processing")  

for token in doc:
    print(token.text, "Vector:", token.has_vector, "OOV:", token.is_oov)
```

```python
doc[0].vector.shape
```

```python
base_token = nlp("bread")
base_token.vector.shape
```

```python
doc = nlp("sandwich burger car tiger human wheat")

for token in doc:
    print(f"{token.text}  <->  {base_token.text}:", token.similarity(base_token))
```

```python
def print_similarity(base_word, words_to_comapre):
    base_token = nlp(base_word)
    doc = nlp(words_to_comapre)
    for token in doc:
        print(f"{token.text} <-> {base_token.text}:", token.similarity(base_token))
```

```python
print_similarity("iphone", "apple samsung iphone dog kitten")
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
women = nlp.vocab["women"].vector
queen =nlp.vocab["queen"].vector

result = king - man + women

```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([result] ,[queen])
```
