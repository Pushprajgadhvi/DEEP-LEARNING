## data augmentation

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from tensorflow.keras.models import load_model
import os
```

```python
img = tf.keras.utils.load_img('../DOGS AND CAT PREDICTION CNN/train/00000-4122619873.png', target_size=(200,200))
img_array = tf.keras.utils.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
```

```python
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,  # random rotation 
    width_shift_range=0.2, # random horizontal  shift
    height_shift_range=0.2, # random vertical shift 
    shear_range=0.2,  # shear transformations
    zoom_range=0.2,  # random zoom
    horizontal_flip=True, # random horizontal flip
    fill_mode='nearest', # fill mode for tranformations
    brightness_range=[0.8, 1.2] # brgihtness range
)
```

```python
augmented_images = datagen.flow(img_array,batch_size=1)  # giving the batch size and the datagen
```

```python
# Generate augmented images using the data generator
augmented_images = datagen.flow(img_array, batch_size=1)  # Create an iterator that applies random transformations to img_array

saved_images = []  # Initialize an empty list to store the filenames of the augmented images

# Loop over the augmented images
for i, batch in enumerate(augmented_images):
    if i >= 5:  # Stop after generating 5 augmented images
        break

    # Convert the batch array back to an image format for saving
    augmented_img = tf.keras.utils.array_to_img(batch[0])

    # Create a filename for the augmented image using the loop index
    filename = f'augmented_image_{i}.png'

    # Save the augmented image to disk
    augmented_img.save(filename)

    # Add the filename to the saved_images list for later use
    saved_images.append(filename)
```
